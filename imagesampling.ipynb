```python
import cv2 as cv
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.neighbors import KernelDensity
```

```python
mnist = tf.keras.datasets.mnist
```

```python
# we have to split into train and test to load the data, but we can just combine X_train and X_test later
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
plt.imshow(X_train[0], cmap='gray')
```

```python
X = np.vstack((X_train, X_test))
```

```python
X_flat = []
for x in X:
    X_flat.append(x.flatten())

X_flat = np.array(X_flat)
X_flat.shape
```

```python
X = X_flat
```

```python
kde = KernelDensity(kernel='gaussian', bandwidth=10).fit(X)
```

```python
def vec_to_matrix(x):
    matrix = []
    for i in range(28):
        row = x[(i * 28):((i + 1) * 28)]
        matrix.append(row)
    return np.array(matrix)
```

```python
n = 100
samples = kde.sample(n)
for i in range(n):
    sample = samples[i]
    img_mat = vec_to_matrix(sample)
    plt.imshow(img_mat, cmap='gray')
    plt.show()
```
